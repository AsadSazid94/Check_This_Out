```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resizing the graph
plt.figure(figsize = (5,3), dpi = 100)

#plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, linestyle = '--',  marker = 'X', markersize = 10, markeredgecolor = 'blue')

plt.plot(x,y, 'g*--', label ='2x') #short hand notation

# 2nd equation
x2 = np.arange(0,4.5,.5)
plt.plot(x2[0:6],x2[0:6]**2, 'r', label = 'x^2')
plt.plot(x2[5:],x2[5:]**2, 'r--')

plt.title('line', fontdict = {'fontname' : 'Times New Roman', 'fontsize' : 40})
plt.xlabel('X axis', fontdict = {'fontname' :'Comic Sans MS', 'fontsize' : 20})
plt.ylabel('Y axis', fontdict = {'fontname' :'Comic Sans MS', 'fontsize' : 20})

plt.xticks([0,1,2,3])
#plt.yticks([0,2,4,6])

plt.legend()

plt.savefig('testing_line_graph.png', dpi = 300)

plt.show()
```
