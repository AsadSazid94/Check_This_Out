### Importing all the necessory library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gas = pd.read_csv('Data/gas_prices.csv')
gas.head()
```

```python
plt.figure(figsize=(8,5)) 

plt.title('Gas Price Over time', fontdict = {'fontweight' : 'bold', 'fontsize' : 20})

plt.plot(gas.Year, gas.USA, 'b.-', label = 'USA')
plt.plot(gas.Year, gas.Canada, 'r*-',  label = 'Canada')
plt.plot(gas.Year, gas['South Korea'], 'gx-',  label = 'South Korea')


for country in gas:
    if country != 'Year':
        #print(country)
        plt.plot(gas.Year, gas[country], marker = '.')

        
# another way to view specific values
#country_to_look_at = ['USA','Canada','South Korea']

#for country in gas:
#    if country in country_to_look_at:
#        plt.plot(gas.Year, gas[country], marker = '.')

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.grid()

plt.legend(gas)

plt.savefig('Gas_price_linegraph.png', dpi = 300)

plt.show()
```

### Reading FIFA data 

```python
fifa = pd.read_csv('Data/fifa_data.csv')
fifa.head()
```

### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins, color = '#50C78B')

plt.xticks(bins)

plt.title('Skill level of players in FIFA')
plt.xlabel('Skill level')
plt.ylabel('No. of players')

plt.show()
```

```python
 fifa.columns
```

### Pie Chart

```python
fifa['Preferred Foot']
```

```python

right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]

plt.style.use('ggplot')

labels = ['right','left']

colors = ['red','blue']

plt.pie([right,left], labels = labels, colors = colors, autopct = '%.2f%%')

plt.title('Prefered foot of FIFA player')

plt.savefig('pie_chart_FIFA_Players_foot.png', dpi = 500)

plt.show()
```

### 2nd Pie chart

```python
fifa_Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
fifa_Weight
#type(fifa_weight[0])

plt.style.use('ggplot')# to get back the default style replace 'ggplot' to 'default'

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['under 125','125-150','150-175','175-200','above 200']
explode = [.4,.4,0,0,.4]
plt.pie(weights, autopct ='%.2f%%', labels = labels, pctdistance = .75 , explode = explode)

#to_kg = 0.45359237
#fifa.Weight * to_kg
plt.title('Weights of fifa players (in lbs)')
plt.show()
```

### 3rd Pie chart

```python
fifa.Club.tail(20)
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
real_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
new_england = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

plt.figure(figsize = (5,8))

labels = ['FC Barcelona', 'Real Madrid', 'England Revolution']

plt.ylabel('FIFA Overall Rating')
plt.title('Football Team Comparision')
# patch_artist = True (to set the face colour)
# medianprops is a dictonary to change the middle line width
boxes = plt.boxplot([barcelona, real_madrid, new_england], labels = labels, patch_artist = True, medianprops = {'linewidth':2}) 

for box in boxes['boxes']:
    box.set(color = 'red', facecolor = 'grey',  linewidth = 2)

#plt.grid()
plt.show()
```
